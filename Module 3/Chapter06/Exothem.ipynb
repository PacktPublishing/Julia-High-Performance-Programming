```julia
function exo(t, x, dx; n=1, a=1, b=1)
  p = x[2]^n * exp(x[1])
  dx[1] = p - a*x[1]
  dx[2] = -b*p
  return(dx)
end;

t = linspace(0.0,5.0,1001);
exo(t, [0.0, 1.0], [0.0, 0.0], a=0.1, b=0.1);

fex(t,x,dx) = exo(t, x, dx, a=0.5, b=0.1);
res = Sundials.cvode(fex, [0.0, 1.0], t);

plot(t,res[:,1])
```
